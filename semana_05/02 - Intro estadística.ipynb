## Histogramas

Una forma de entender los datos es contar cuántas veces aparece cada valor en el [sample](https://en.wikipedia.org/wiki/Sample_(statistics)#:~:text=In%20statistics%20and%20quantitative%20research,points%2C%20sampling%20units%20or%20observations.).

```python
sample = [1,2,3,4,5,6]
```

```python
# Experimento: tirar un dado 20 veces
exp = random.choices(sample, k=20)
```

```python
# Hacemos una comp dict
count_ = {n:list(exp).count(n) for n in exp}
```

```python
count_
```

```python
# built-in alternativa
from collections import Counter
```

```python
#Counter hace esto mismo
Counter(exp)
```

```python
for k,v in Counter(exp).items():
    print (f"{k}---> {v}")
```

Y podríamos trazar el resultado como un gráfico de barras para tener una representación visual.

```python
plt.bar(x=count_.keys(),
        height=count_.values())
plt.xlabel('Número dado')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# seaborn tiene una buena función para hacer esto también que ya conocemos de sobra
sns.countplot(x=exp, color='b');
```

Observamos que el método anterior sólo es válido para [variables discretas](http://onlinestatbook.com/glossary/discrete_variables.html). La librería pandas tiene un método `pd.cut` que sirve exactamente para ese propósito. 

```python
binned = pd.cut(exp, bins=6)
print(binned.value_counts())
```

Las tablas anteriores se denominan "tablas de frecuencia". Una representación visual de una tabla de frecuencias es lo que llamamos un [histograma](https://en.wikipedia.org/wiki/Histogram)

```python
# La diferencia con density es que me pinta la distribución de la probabilidad poniendo density
plt.hist(exp, bins="auto", density=True)# Elegir el número de bins es engañoso
# Density me cambia el valor numérico por el porcentaje
plt.show()
```

De lo visto hasta ahora, observamos:
* Si se da un entero, se calculan y devuelven los bordes de los bins + 1, de forma consistente con `numpy.histogram.`

* Todo menos el último bin (el más a la derecha) está medio abierto. En otras palabras, si bins es `[1, 2, 3, 4]`
entonces el primer bin es `[1, 2)` (incluyendo 1, pero excluyendo 2) y el segundo `[2, 3)`. 
Sin embargo, el último bin es `[3, 4]`, que incluye el 4

### Varianza y desviación estándar

Además de saber dónde está *centrada* una distribución, también es importante saber lo *variable* que son los datos, es decir, lo *dispersos* que están. En este caso son útiles dos estadísticas, la varianza y la desviación estándar:

$$Var = \sigma^2 = \frac{\sum (x_i - \mu) ^ 2}{N}$$ 

$$Std = \sigma = \sqrt{Var}$$

De [Wikipedia:](https://en.wikipedia.org/wiki/Variance) En teoría de la probabilidad y estadística, la varianza es la expectativa de la desviación al cuadrado de una variable aleatoria respecto a su media. Informalmente, mide la distancia de un conjunto de números con respecto a su valor medio. 

(**Nota**: Las expresiones anteriores se refieren a la Varianza de toda una población, donde $\mu$ es la media de la población. Cuando estimamos la Varianza o la std a partir de una muestra, la fórmula correcta e *insesgada* es:
$$\hat{Var} = \hat\sigma^2 =\frac{\sum (x_i - \bar{x}) ^ 2}{N-1}$$

### Covarianza

[Wikipedia:](https://en.wikipedia.org/wiki/Covariance)

En teoría de la probabilidad y estadística, la covarianza es una medida de la variabilidad conjunta de dos variables aleatorias[1] Si los valores mayores de una variable se corresponden principalmente con los valores mayores de la otra variable, y lo mismo ocurre con los valores menores, (es decir, las variables tienden a mostrar un comportamiento similar), la covarianza es positiva[2] En el caso contrario, cuando los valores mayores de una variable se corresponden principalmente con los valores menores de la otra, (es decir, las variables tienden a mostrar un comportamiento opuesto), la covarianza es negativa.       
**El signo de la covarianza muestra, por tanto, la tendencia en la relación lineal entre las variables**. 




La fórmula de la covarianza entre dos varaibles ($j$, y $k$) es similar a la de la varianza:

$$q_{jk}=\frac{1}{N-1}\sum_{i=1}^{N}\left(  x_{ij}-\bar{x}_j \right)  \left( x_{ik}-\bar{x}_k \right)$$

**Nota:** ¿puedes ver que si $j=k$ la covarianza es sólo la varianza?

### Correlación de Pearson

La magnitud de la covarianza no es fácil de interpretar porque no está normalizada y, por tanto, depende de las magnitudes de las variables. Sin embargo, la versión normalizada de la covarianza, el coeficiente de correlación de Pearson, muestra por su magnitud la fuerza de la relación lineal. La fórmula es

$$\rho_{X,Y} = \frac{Cov(X,Y)}{\sigma_X \sigma_Y}$$


Donde $\sigma_X$ y $\sigma_Y$ son las respectivas desviaciones estándar.

El coeficiente de correlación de Pearson es una estadística que mide la correlación lineal entre dos variables X e Y. Tiene un valor entre +1 y -1. Un valor de +1 es una correlación lineal positiva total, 0 es ninguna correlación lineal y -1 es una correlación lineal negativa total.


**¡CUIDADO!** El hecho de que la correlación (o covarianza) sea cero no significa que no haya patrones

Un coeficiente de correlación alternativo conocido como [coeficiente de correlación de Spearman](https://en.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient) resuelve este problema calculando la correlación de Pearson para los rangos. El signo de la correlación de Spearman indica la dirección de la asociación entre X (la variable independiente) e Y (la variable dependiente). Si Y tiende a aumentar cuando X aumenta, el coeficiente de correlación de Spearman es positivo. Si Y tiende a disminuir cuando X aumenta, el coeficiente de correlación de Spearman es negativo. Una correlación de Spearman de cero indica que no hay tendencia a que Y aumente o disminuya cuando X aumenta. La correlación de Spearman aumenta en magnitud a medida que X e Y se acercan a ser funciones perfectamente monótonas entre sí. Cuando X e Y están perfectamente relacionados de forma monótona, el coeficiente de correlación de Spearman se convierte en 1.

Para más información, consulte [este post](https://towardsdatascience.com/clearly-explained-pearson-v-s-spearman-correlation-coefficient-ada2f473b8) en towardsdatascience.
